```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pydotplus
import re, string
%matplotlib inline
from sklearn import tree
from IPython.display import Image
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('sms_raw.csv')
```

```python
sms = df[['spam', 'message']].copy()
```

```python
sms.spam.value_counts() #See class imbalance
```

```python
sms['words'] = sms.message.str.lower().str.split()
sms['msg_length'] = sms.words.apply(len)
print(sms[sms.spam == 'ham'].msg_length.mean()) #Nice difference in mean msg length
print(sms[sms.spam == 'spam'].msg_length.mean())
```

```python
def strip_punc(words):
    """Takes a list of words and returns a list of words
    without punctuation
    """
    translator = str.maketrans('', '', string.punctuation)
    result = []
    for word in words:
        result.append(word.translate(translator))
    return result

sms.words = sms.words.apply(strip_punc)
```

```python
def get_words(series):
    """Takes a series of lists of words and returns
    a single series of all words
    """
    words = []
    for item in series:
        words += item
    return pd.Series(words)

spam_words = get_words(sms[sms.spam=='spam'].words)
ham_words = get_words(sms[sms.spam=='ham'].words)
```

```python
spam_words = ['call', '2', 'free', 'txt', 'now']
sms['spammy_words'] = False
for row in sms.index: 
    for word in spam_words:
        if word in sms.loc[row].words:
            sms.set_value(row, 'spammy_words', True)   #Flag all those spammy messages
```

```python
sms['colon'] = False
for row in sms.index:
    sms.set_value(row, 'colon', bool(re.match('.*:', sms.loc[row].message)))
```

```python
sms = sms.sample(frac=1)
```

```python
#Now I have some features, gogo gadget decision tree!
features = ['msg_length', 'spammy_words', 'colon']
X=sms[features]
Y=sms.spam

dec_tree = tree.DecisionTreeClassifier(
    criterion='entropy'
    #max_depth=3
    #class_weights='balanced'
)

dec_tree.fit(X, Y)

dot_data = tree.export_graphviz(
    dec_tree, 
    out_file=None,
    feature_names=features,
    class_names=['ham', 'spam'],
    filled=True
)

graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
scores = cross_val_score(dec_tree, X, Y, cv=10)
pd.Series(scores).mean()
```
